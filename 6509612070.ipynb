```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

choose column

```python
dataset = pd.read_csv('/content/drive/MyDrive/Data/Titanic.csv',usecols=[1,2,4,5,6])
dataset
```

```python
dataset.head()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset = pd.get_dummies(dataset,columns=['pclass','sex'])
dataset
```

data.iloc[row,column]

```python
X_train, X_test, y_train, y_test =train_test_split(dataset.iloc[:,1:],dataset.iloc[:,0],test_size=0.2,random_state =21)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
```

MLP has random_state

note              alpha          train        test
    adam          0.1         0.77        0.80
                 0.005       0.77        0.81


```python
clf=MLPClassifier(hidden_layer_sizes=2,activation='logistic',solver='adam',random_state=21,alpha=0.01)
clf.fit(X_train,y_train)
print("Accuracy for training: {}".format(clf.score(X_train,y_train)))
print("Accuracy for testing: {}".format(clf.score(X_test,y_test)))
```

```python
y_predict_train = clf.predict(X_train)
y_predict_train
```

```python
y_predict_test = clf.predict(X_test)
y_predict_test
```

```python
matrix_train = confusion_matrix(y_train,y_predict_train)
matrix_train
```
